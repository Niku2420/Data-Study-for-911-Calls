```python
dataFrame['SpecificReasons']=dataFrame['title'].apply(lambda title: title.split(':')[1])
```

```python
dataFrame['BroadReasons']=dataFrame['title'].apply(lambda title: title.split(':')[0] )
```

```python
sns.countplot(x='BroadReasons',data=dataFrame )
```

```python
#now converting of timestamp from string value to to actual timestamp value
dataFrame['timeStamp']=pd.to_datetime(dataFrame['timeStamp'])
```

```python
#to check wether the timestamp is changed or not 
dataFrame.info()
```

```python
dataFrame['timeStamp'].head(10)
```

```python
type(dataFrame['timeStamp'].iloc[0])
```

```python
dataFrame['timeStamp'].iloc[0]
```

```python
# for creating column for hour
dataFrame['Hour']=dataFrame['timeStamp'].apply(lambda time: time.hour)
```

```python
dataFrame.head(10)
```

```python
# for creating column for months
dataFrame['Months']=dataFrame['timeStamp'].apply(lambda time: time.month)
```

```python
dataFrame.head()
```

```python
dataFrame['Day']=dataFrame['timeStamp'].apply(lambda time: time.dayofweek)
```

```python
dataFrame.head(20)
```

```python
dataFrame['Day'].unique()# with unique we can count unique values present in big data
```

```python
sns.countplot(x='Months',data=dataFrame,hue='BroadReasons')
```

```python
# using of groupby method of seaborn
byMonth=dataFrame.groupby('Months').count() 
```

```python
byMonth
```

```python
byMonth['twp'].plot() #according to months (survey)
```
